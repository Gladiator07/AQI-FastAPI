```python
import pandas as pd
import json
from datetime import datetime
from tqdm import tqdm
```

```python
f = open('data.json', 'r')
data = json.load(f)

rows = []
for i in range((len(data["list"]))):
    
    data["list"][i]["dt"] = datetime.utcfromtimestamp(data["list"][i]["dt"]).strftime('%Y-%m-%d %H')
    air_contents = data["list"][i]["components"]
    
    date = data["list"][i]["dt"].split(' ')[0]
    time = data["list"][i]["dt"].split(' ')[1]
    AQI = data["list"][i]["main"]["aqi"]
    co = air_contents["co"]
    no = air_contents["no"]
    no2 = air_contents["no2"]
    o3 = air_contents["o3"]
    so2 = air_contents["so2"]
    pm2_5 = air_contents["pm2_5"]
    pm10 = air_contents["pm10"]
    nh3 = air_contents["nh3"]
    
    rows.append([date,time,AQI,co,no,no2,o3,so2,pm2_5,pm10,nh3])
df = pd.DataFrame(rows, columns=["date","time","AQI","co","no","no2","o3","so2","pm2_5","pm10","nh3"])

df.head()
```
